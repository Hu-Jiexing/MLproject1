```python
import os 
import numpy as np
import pandas as pd
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import StratifiedKFold as sk
from sklearn.tree import DecisionTreeClassifier as dt
from sklearn.metrics import accuracy_score as acs
from sklearn import tree

from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import time
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_graphviz

# Load dataset
df = pd.read_csv("cleaneddata.csv")

display(df)

# Extract independent and dependent variables
X = df.drop('Condition', axis=1)
y = df['Condition']

# create save path
# os.getcwd() returns current working directory
path = os.getcwd()
savePath = path + '/decision_tree_results/'
print(savePath)
```

```python
# Decision Tree

# Define decision model
model_name = 'Decision Tree'
dt_model1 = dt(random_state=0)

# Split data into 3/4 for training and 1/4 for testing
cv = sk(n_splits=4, shuffle=True, random_state=0)
# Accuracy list
accuracy_l1 = []
# Result list
result_matrix =[]

for i, (tr_i, te_i) in enumerate(cv.split(X, y)):
    X_tr = X.iloc[tr_i]
    X_v = X.iloc[te_i]
    y_tr = y.iloc[tr_i]
    y_v = y.iloc[te_i]

    #print("Training set:", X_tr)
    #print("Validation set:", X_v)

    print("\ntraining data length:", len(y_tr))
    print("verification data length:", len(y_v))
    
    # Fit decision tree model and predict y_p
    dt_model1.fit(X_tr, y_tr)
    y_p = dt_model1.predict(X_v)

    print(f'\nSplit {i+1} metrics:')

    # Calculate accuracy
    accuracy = np.mean(acs(y_v, y_p))
    print("Accuracy:", accuracy.round(4))
    accuracy_l1.append(accuracy)

    # Mean accuracy score
    m_sc = round(np.mean(accuracy_l1), 4)
    print('%s score :  %s' % (model_name, "{:.4f}".format(m_sc)))

    # Confusion matrix and classification report
    a = confusion_matrix(y_v,y_p)
    print("Decision tree Prediction confusion_matrix", "\n", a)
    print("\n")

    print("Decision tree Prediction classification report", "\n", classification_report(y_v,y_p))

    cond1_accuracy =  a[0,0] / (a[0,1]+a[0,2]+a[0,0])
    print("Condition1 accuracy:", cond1_accuracy.round(2))
                                            
    cond3_accuracy = a[1,1] / (a[1,0] + a[1,1] + a[1,2])
    print("Condition3 accuracy:", cond3_accuracy.round(2))

    normal1_accuracy = a[2,2] / (a[2,0] + a[2,1] + a[2,2])
    print("Normal1 accuracy:", normal1_accuracy.round(2))

    # Use confusion matrix to compare y_v and y_p, plot and save confusion matrix
    fig, ax = plt.subplots(figsize=(5,5))
    
    # Use heatmap
    ax = sns.heatmap(confusion_matrix(y_v, y_p), annot=True, fmt='d', cmap='Blues', xticklabels=dt_model1.classes_, yticklabels=dt_model1.classes_)
    plt.tight_layout()
    #plt.savefig(savePath + '/' + 'dt_model1_' + str(len(accuracy_l1)) + 'confusion matrix.png')

    fig = plt.figure(figsize=(15, 15))
    # Plot and save Decision tree 
    tree.plot_tree(dt_model1, feature_names=X.columns,  class_names=dt_model1.classes_, filled=True)
    #plt.savefig(savePath + '/' + 'dt_model1_' + str(len(accuracy_l1)) + 'decision_tree.png')
    plt.show()
    
    # Reset the label to original and save the original test data +prediction into a .csv file 
    X_v.reset_index(drop=True, inplace=True)
    y_v.reset_index(drop=True,inplace=True)
    result_matrix = pd.concat([X_v, y_v], axis=1,ignore_index=True)
    result_matrix.loc[:,13] = y_p 
    result_matrix.to_csv(savePath + '/' + 'cv1_' + str(len(accuracy_l1)) + '.csv', index=False)
```

```python
# Identifying Important Features in Decision Trees
# Calculate feature importances

importances = dt_model1.feature_importances_

# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]
#print("indices:", indices)

# Rearrange feature names so they match the sorted feature importances
names = [X.columns[i] for i in indices]
print("Feature names:", names)
print("Feature importance:", importances[indices])

# Cumulative explained variance
cumulative_importance = np.round(np.cumsum(importances), decimals=3)
print("\nCumulative Explained Variance (%):")
print(cumulative_importance)

#Create plot
plt.figure()

# Create plot title
plt.title("Feature Importance")

# Add bars
bars = plt.bar(range(X.shape[1]), importances[indices])

# Add feature names as x-axis labels
plt.xticks(range(X.shape[1]), names, rotation=90)

# Add value labels on top of each bar
for bar in bars:
    yval = bar.get_height()  # Get the height of each bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 3), 
             ha='center', va='bottom', fontsize = 9)  # Add text above the bar

# Show plot
plt.show()
```

```python
# RFE for decision tree
# Specify upper limit of features used for test
test_components = 12

# Initialize the decision tree estimator
dt_estimator = dt(random_state=0)

acc_list = []
max_list = []
min_list = []

print("Recursive Feature Selection With:")

# Loop through increasing numbers of features to test (from 1 to 12)
for n_components in range(1, test_components + 1):
    print(f"\n{n_components} Selected Feature(s)")

    fold_acc = []

    # Initialize RFE (Recursive Feature Elimination) with the estimator
    # n_features_to_select=2 means that it will select n features
    selector = RFE(dt_estimator, n_features_to_select=n_components)

    for i, (tr_i, te_i) in enumerate(cv.split(X, y)):
        X_tr = X.iloc[tr_i]
        X_v = X.iloc[te_i]
        y_tr = y.iloc[tr_i]
        y_v = y.iloc[te_i]

        # Fit RFE to the training data and transform X_train to select the top features
        X_tr_selected = selector.fit_transform(X_tr, y_tr)

        # Train the decision tree model using the selected features
        dt_estimator.fit(X_tr_selected, y_tr)

        # Transform the validation set to the selected features
        X_v_selected = selector.transform(X_v)

        # Predict on the test set
        y_pred_rfe_n = dt_estimator.predict(X_v_selected)

        # Evaluate the model
        accuracy = np.mean(acs(y_v, y_pred_rfe_n))
        fold_acc.append(accuracy)
        print(f"Split {i+1} Accuracy:", round(accuracy,4))
    
    avg_accuracy = np.mean(fold_acc)
    acc_list.append(avg_accuracy)
    max_list.append(max(fold_acc))
    min_list.append(min(fold_acc))

    # Print the performance metrics for the current number of features
    print(f"Average Accuracy: {round(avg_accuracy,4)}")

    
```

```python
# RMSE and RÂ² for different numbers of components
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot max accuracy
ax1.set_xlabel('Number of Features')
ax1.set_ylabel('Accuracy')
ax1.plot(range(1, test_components + 1), max_list, color='tab:red', marker='o', label='max')
ax1.plot(range(1, test_components + 1), min_list, color='tab:blue', marker='x', label='min')

plt.title('Accuracy for Different Number of Features With K-fold')
fig.tight_layout()
plt.show()
```

```python
# RMSE and RÂ² for different numbers of components
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot max accuracy
ax1.set_xlabel('Number of Features')
ax1.set_ylabel('Accuracy')
ax1.plot(range(1, test_components + 1), max_list, color='tab:red', marker='o', label='max')
ax1.plot(range(1, test_components + 1), min_list, color='tab:blue', marker='x', label='min')

plt.title('Accuracy for Different Number of Features With K-fold')
fig.tight_layout()
plt.show()
```
